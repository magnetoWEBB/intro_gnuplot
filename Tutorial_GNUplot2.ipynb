### 7.4. COLORES Y ANCHOS DE LÍNEA

Podemos controlar el color y ancho de líneas con <b> linecolor "color"</b> y con <b> linewidth "número"</b>
    
Los colores se especifican de acuerdo a un número: 1, 2, 3, 4, 5, etc.    
Los anchos se especifican de manera usual: 1, 2, 3, 4.

```gnuplot
reset 

set title "Función y=e^{|x|}" font ",14"

set xrange[-4:4]
set yrange[0:50]

set grid #Para ubicar la grilla

set key top center font ",14" #Noten que también cambiamos la localización del key.

set tics font ",14"
set xtics -4,2,4
set ytics 0,10,50

set xlabel "Eje X" font ",14" offset -6,+1.5
set ylabel "Eje Y" font ",14" rotate by 0 offset +1.5,0

plot exp(abs(x)) title " y = e^{|x|}" linecolor 2 linewidth 3
```

También funciona con los diminutivos <b>lc</b> y <b>lw</b> de <b>linecolor "color"</b> y con <b>linewidth "número"</b>, respectivamente.

```gnuplot
reset 

set title "Función y=e^{|x|}" font ",14"

set xrange[-4:4]
set yrange[0:50]

set grid #Para ubicar la grilla

set key top center font ",14" #Noten que también cambiamos la localización del key.

set tics font ",14"
set xtics -4,2,4
set ytics 0,10,50

set xlabel "Eje X" font ",14" offset -6,+1.5
set ylabel "Eje Y" font ",14" rotate by 0 offset +1.5,0

plot exp(abs(x)) title " y = e^{|x|}" lc 2 lw 3
```

Además, podemos utilizar <b>dashtype número</b> para líneas entrecortadas:

```gnuplot
reset 

set title "Función y=e^{|x|}" font ",14"

set xrange[-4:4]
set yrange[0:50]

set key top center font ",14"

set tics font ",14" #Modifica el tamaño de letra de los ejes. En este caso de los números.
set xtics -4,2,4
set ytics 0,10,50

set xlabel "{/Symbol b}" font ",14" offset -7,+1.5
set ylabel "{/Symbol e}" font ",14" rotate by -1 offset +1.5,0

plot exp(abs(x)) title " y = e^{|x|}" linecolor 4 linewidth 4 dashtype 3, 5*x**2 title " y = x^2" linecolor 3 linewidth 5 dashtype 4
```

## 8. Lectura de datos

En forma general, utilizamos el comando <b>plot</b> seguido por "<b>el nombre del archivo entre comillas</b>" y el comando <b>using</b> para indicar el orden de las columnas.

El comando <b>plot</b> asume que el fichero es un archivo de texto, separado por espacios o "tabs".

### 8.1. Ficheros de texto:

```gnuplot
reset

plot "ejemplo_datos_1.dat" using 1:2

#plot "ejemplo_datos_1.dat" using 2:1
```

Noten que por defecto plotea los datos con "points". Podemos usar </b>with lines</b> para graficar con líneas.

```gnuplot
reset

plot "ejemplo_datos_1.dat" using 1:2 with lines
```

Otra alternativa para visualizar los datos como líneas es utilizar el comando:

<b>set style data lines</b>

Que, de hecho, es más conveniente.

Otros estilos incluyen:

```gnuplot
help plotting styles
```

Por ejemplo, tomemos otro de los archivos de ejemplo:

        ejemplo_datos_0.dat


```gnuplot
reset 

set style data points

plot "ejemplo_datos_0.dat" using 1:2
```

Para graficar con ambos: puntos y líneas, usamos <b>linespoints</b> (el diminutivo es <b>lp</b>):

```gnuplot
reset 

set style data linespoints #Para superponer puntos y líneas

plot "ejemplo_datos_0.dat" using 1:2 linecolor 2
```

Otra alternativa para combinar líneas y puntos de un mismo conjunto de datos es utilizar:

        set style data lines
        
Como sigue a continuación:

```gnuplot
reset 

set style data lines

plot "ejemplo_datos_0.dat" using 1:2 with lines linecolor 2 linewidth 2, "ejemplo_datos_0.dat" using 1:2 with points pointtype 1 pointsize 1 linecolor rgb "red" notitle

#Noten que también podemos utilizar linecolor rgb (código red green blue) para definir el color.
```

El comando <b>using</b> puede usarse para realizar operaciones con las columnas del fichero fuente.

(\$1/2):(\$2*3)

```gnuplot
reset 

set style data lines

plot "ejemplo_datos_0.dat" using ($1/2):2 with lines linecolor 2 linewidth 2, "ejemplo_datos_0.dat" using ($1/2):2 with points pointtype 1 pointsize 1 linecolor rgb "red" notitle

#Noten que también podemos utilizar linecolor rgb (código red green blue) para definir el color.
```

Además, podemos utilizar escala logarítmica con el comando:

        set logscale x/y

En la mayoría de casos resulta conveniente definir los estilos de línea antes de usarlos, e.g.:

        set style line 1 lt 1 lw 3.0 lc rgb "dark-red"
        
Luego, podmos invocar as estilo con 

        linestyle 1

O con su diminutivo:

        ls 1

En la curva deseada.

```gnuplot
reset

#Definimos los estilos de línea con código RGB
set style line 1 lt 1 lw 3.0 lc rgb "dark-red"
set style line 2 lt 1 lw 3.0 lc rgb "#E9967A"
set style line 3 lt 1 lw 3.0 lc rgb "#228B22"

#definimos el borde
set border linewidth 3

#definimos el estilo
set style data lines

#ubicamos una etiqueta
set label "PDFs" at -10,2.46 tc rgb "black"

#definimos los rangos
set xrange [-10:10]
set yrange [-7:2]

#etiquetamos los sjes
set xlabel offset +0.0, +0.65
set xlabel "ln({/Symbol r}/{/Symbol n})"
set ylabel offset +0.5
set ylabel "log_{10}P({/Symbol r}/{/Symbol n})"

#definimos el espaciamiento para los ejes
set xtics 2
set ytics -7,1,0

#Definimos la posición y largo del "key"
set key left top
set key samplen 10

plot "pdf_f00.dat" using 1:2 title "t/t_{cc}=0.00" linestyle 1, "pdf_f15.dat" using 1:2 title "t/t_{cc}=0.25" linestyle 2, "pdf_f30.dat" using 1:2 title "t/t_{cc}=0.50" linestyle 3
```

    set bmargin at screen NÚMERO(0-1)
    set tmargin at screen NÚMERO(0-1)

    set rmargin at screen NÚMERO(0-1)

```gnuplot
reset

#Definimos los estilos de línea con código RGB
set style line 1 lt 1 lw 3.0 lc rgb "dark-red"
set style line 2 lt 1 lw 3.0 lc rgb "#E9967A"
set style line 3 lt 1 lw 3.0 lc rgb "#228B22"

#definimos el borde
set border linewidth 3

#definimos el estilo
set style data lines

#ubicamos una etiqueta
set label "PDFs" at -10,2.46 tc rgb "black"

#definimos los rangos
set xrange [-10:10]
set yrange [-7:2]

#etiquetamos los sjes
set xlabel offset +0.0, +0.65
set xlabel "ln({/Symbol r}/{/Symbol n})"
set ylabel offset +0.5
set ylabel "log_{10}P({/Symbol r}/{/Symbol n})"

#definimos el espaciamiento para los ejes
set xtics 2
set ytics -7,1,0

#Definimos la posición y largo del "key"
set key left top
set key samplen 10

#Definimos los márgenes
set bmargin at screen 0.14
set tmargin at screen 0.92
set lmargin at screen 0.13
set rmargin at screen 0.94

plot "pdf_f00.dat" using 1:2 title "t/t_{cc}=0.00" linestyle 1, "pdf_f15.dat" using 1:2 title "t/t_{cc}=0.25" linestyle 2, "pdf_f30.dat" using 1:2 title "t/t_{cc}=0.50" linestyle 3
```

### BARRAS DE ERROR

Para ubicar barras de error, utilizamos <b>with xyerrorbars</b>:
    
        plot 'fichero.dat' using 1:2:(sqrt($1)):(sqrt($2)) with xyerrorbars
        
En el ejemplo a continuación tomamos el archivo: ejemplo_datos_5.dat

Y tomamos los errores como la raíz cuadrada de las columnas 1 y 2.

```gnuplot
reset

plot 'ejemplo_datos_5.dat' using 1:2:(sqrt($1)):(sqrt($2)) with xyerrorbars
```

El comando <b>with xyerrorbars</b> acepta hasta 6 argumentos para el comando precedente <b>using</b>, como sigue:


      using COLUMNA_X:COLUMNA_Y:ERROR_INFERIOR_X:ERROR_SUPERIOR_X:ERROR_INFERIOR_Y:ERROR_SUPERIOR_Y with xyerrorbars

```gnuplot
reset
set key tmargin left
set datafile separator "," #Para abrir archivos .csv

plot 'ejemplo_datos_6.csv' using 1:2:($1-$5):($1+$6) with xerrorbars
```

```gnuplot
reset
set key tmargin left
set datafile separator "," #Para abrir archivos .csv

plot 'ejemplo_datos_6.csv' using 1:2:($2-$7):($2+$8) with yerrorbars
```

## 9. Ajustes de curvas con GNUplot

```gnuplot
reset

f(x) = a*x + b

fit f(x) 'ejemplo_datos_0.dat' using 1:2 via a,b

plot 'ejemplo_datos_0.dat' using 1:2, f(x)
```

```gnuplot
reset

g(x) = a*sqrt(x) + b

f(x) = c*(x**4) + d*(x**3) + e*(x**2) + j*(x) + k

fit g(x) 'ejemplo_datos_0.dat' using 2:4 via a,b

fit f(x) 'ejemplo_datos_0.dat' using 2:4 via c,d,e,j,k

plot 'ejemplo_datos_0.dat' using 2:4, g(x),f(x)
```

```gnuplot
reset

f(x) = a*x**5 + b*x**4 + c*x**3 + d*x**2 + e*x

fit f(x) 'ejemplo_datos_0.dat' using 1:2 via a,b,c,d,e

plot 'ejemplo_datos_0.dat' using 1:2, f(x)
```

```gnuplot
reset

f(x) = a*x**5 + b*x**4 + c*x**3 + d*x**2 + e*x

fit f(x) 'ejemplo_datos_0.dat' using 1:2 via a,b,c,d,e

plot 'ejemplo_datos_0.dat' using 1:2:(sqrt($10/200)):(sqrt($3/400)) with xyerrorbars, f(x)
```
