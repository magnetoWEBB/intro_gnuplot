Algunas caaractaerísticas de GNUplot se enuncian a continuación:

- Gnuplot es una herramienta de ploteo basada en comandos tomados de los lenguajes C y Fortran.


- Funciona en múltiples plataformas: Unix, Linux, MacOSX y Windows.


- Gnuplot es gratis y de libre distribución.


- Es la librería de graficación por defecto de sistemas GNU ("GNU's Not Unix").


- La primera versión data del año 1986.


- Fue creado con el fin de facilitar el ploteo de funciones matemáticas y la producción de imágenes de alta calidad.


- Es ampliamente utilizado en la comunidad científica.


- Tiene una interfaz interactiva y amigable con el usuario.


- Octave (la versión libre de Matlab) utiliza a GNUplot como su motor de graficación.

## 3. Gráficos básicos en GNUplot

## 5. Gráficas de superficies:

Para gráficas de superficies de la forma $z=z(x,y)$, podemos utilizar el comando $\textbf{splot}$

```gnuplot
help splot
```

Por ejemplo: $z=\frac{xy}{30}$

```gnuplot
splot sin(x*y/30)
```

Otro ejemplo: $z=\frac{1}{x^2}+\frac{1}{y^2}$

```gnuplot
splot ((1/(x**2)+(1/(y**2))))
```

Al igual que los plots anteriores, también podemos graficar múltiples funciones con $\textbf{plot}$ utilizando comas para separarlas. Por ejemplo, grafiquemos las funciones:

$y=f(x,y)=x^2+y^2$,

y también:

$y=f(x,y)=x^2-y^2$

```gnuplot
splot x**2+y**2, x**2-y**2
```

## 6: Gráficos paramétricos

Para gráficos paramétricos, debemos usar el comando $\textbf{set}$ y el argumento $\textbf{parametric}$ como sigue:

```gnuplot
set parametric
```

Podemos plotear curvas con $\textbf{t}$ como argumento:

Por ejemplo, grafiquemos la función hélice $f(t)$, definida por:

$x(t)=r\,\cos(\omega\,t)$

$y(t)=r\,\epsilon\sin(\omega\,t)$

$z(t)=k\,t$

Donde $r$ es el radio de la espiral, $\omega$ es el ángulo girado por unidad de tiempo $t$, $k$ es el avance en $z$ por unidad de tiempo $t$, y $\epsilon=\pm 1$ da el sentido de giro.

```gnuplot
splot [t=-40:40] (2*cos(0.5*t)), (3*sin(0.5*t)), (10*t)
```

Cambiemos la dirección de giro de la hélice:

```gnuplot
splot [t=-50:50] (2*cos(0.5*t)), (-5*sin(0.5*t)), (2*t)
```

Otros ejemplos de curvas paramétricas usando $\textbf{plot}$ en lugar de $\textbf{splot}$ se presentan a continuación:

Grafiquemos una función espiral paramétrica:

$x(t)=t\,\sin(t)$

$y(t)=t\,\cos(t)$

```gnuplot
plot [0:10*pi] t*sin(t),t*cos(t)
```

Y ahora grafiquemos una función cardioide:

$x(t)=\sin(t)(1-\cos(t))$
$y(t)=\cos(t)(1-\cos(t))$

```gnuplot
plot [0:2*pi] sin(t)*(1-cos(t)), cos(t)*(1-cos(t))
```

También podemos plotear superficies.

Por ejemplo:

$x(u,v)=\cos(u)\,\cos(v)$

$y(u,v)=\cos(u)\,\sin(v)$

$z(u,v)=\sin(u)$

Donde $u$ y $v$ tienen rangos definidos.

```gnuplot
splot [-pi/2:pi/2] [0:2*pi] cos(u)*cos(v),cos(u)*sin(v),sin(u)
```

O un torus:

$x(u,v)=(1-0.1\,\cos(v))\,\cos(u)$

$y(u,v)=(1-0.1\,\cos(v))\,\sin(u)$

$z(u,v)=0.3\,\sin(v)$

```gnuplot
splot [0:2*pi] [0:2*pi] (1-0.1*cos(v))*cos(u),(1-0.1*cos(v))*sin(u),0.3*sin(v)
```

Para regresar al modo normal, revertimos la opción paramétrica con $\textbf{unset}$:

```gnuplot
unset parametric
```

Y graficamos otra curva no paramétrica, por ejemplo:

$\left(x^2+y^2-1\right)^3=x^2y^3$

```gnuplot
splot (x**2+y**2-1)
```

## 7. Personalización de Gráficos

### 7.1. RANGOS Y ESPACIAMIENTO:

Veamos un ejemplo con $\textbf{plot}$:

$y=e^{\,|\,x\,|}$

```gnuplot
plot exp(abs(x))
```

Cambiemos el rango de $x$ e $y$:

```gnuplot
plot [x=-5:5] [y=0:100] exp(abs(x))
```

Otra forma es a través del comando $\textbf{set x/yrange[a:b]}$

```gnuplot
set xrange[-4:4]
```

```gnuplot
set yrange[0:50]
```

```gnuplot
plot exp(abs(x))
```

También podemos cambiar el espaciamiento sobre los ejes con $\textbf{set x/ytics a,c,b}$, donde $\textbf{a}$, y $\textbf{b}$ son los extremos y $\textbf{c}$ es el espaciamento.

```gnuplot
set xtics -4,2,4
```

```gnuplot
set ytics 0,25,50
```

```gnuplot
plot exp(abs(x))
```

También podemos limpiar todos los comandos introducidos por $\textbf{set}$ a través del comando $\textbf{reset}$:

```gnuplot
reset
```

Ahora tomemos el último ejemplo con $\textbf{splot}$ y cambiemos el rango de $x$ e $y$:

```gnuplot
splot [x=-20:20] [y=-20:20] (x**2+y**2-1)
```

También podemos controlar el rango de $z$:

```gnuplot
splot [x=-20:20] [y=-20:20] [z=0:1500] (x**2+y**2-1)
```

Una segunda forma de modificar los rangos es a través de $\textbf{set x/y/zrange}$

```gnuplot
set xrange [-40:40]
```

```gnuplot
set yrange [-40:40]
```

```gnuplot
set zrange [0:1500]
```

Y también podemos cambiar el espaciamiento sobre los ejes:

```gnuplot
set xtics -40,20,40
```

```gnuplot
set ytics -40,40,40
```

```gnuplot
set ztics 0,300,1500
```

```gnuplot
splot (x**2+y**2-1)
```

Podemos usar el comando $\textbf{unset}$ para remover uno o más de las propiedades del gráfico:

```gnuplot
unset ztics
```

```gnuplot
splot (x**2+y**2-1)
```

```gnuplot
reset
```

```gnuplot
splot (x**2+y**2-1)
```

Noten que al usar $\textbf{reset}$ regresamos al gráfico original, sin ninguna personalización.

### 7.2. TÍTULOS, ETIQUETAS, Y EJES

Tomemos el ejemplo anterios con $\textbf{plot}$:

$y=e^{\,|\,x\,|}$

```gnuplot
reset 

set xrange[-4:4]
set yrange[0:50]

set xtics -4,2,4
set ytics 0,10,50

plot exp(abs(x))
```

Podemos asignar un título al gráfico con $\textbf{set title}$:

```gnuplot
set title "Función e^x"
```

También podemos usar $\textbf{replot}$ para repetir el último plot

```gnuplot
replot
```

Podemos cambiar la ubicación de la leyenda:

```gnuplot
set key left bottom
```

```gnuplot
replot
```

Podemos retirar la leyenda con $\textbf{unset}$

```gnuplot
unset key
```

```gnuplot
replot
```

Podemos también modificar el nombre de los ejes con set $\textbf{x/y/zlabel}$:

```gnuplot
set xlabel "Eje X"
set ylabel "Eje Y"
```

```gnuplot
replot
```

Ahora un ejemplo con $\textbf{splot}$:

```gnuplot
reset

set key bmargin right # Abajo del margen - Éste es un comentario

set parametric

set title "Donas entrecruzadas"

set urange [-pi:pi]
set vrange [-pi:pi]

#Podemos poner títulos a las leyendas con title "título"

splot cos(u)+0.5*cos(u)*cos(v),sin(u)+0.5*sin(u)*cos(v),0.5*sin(v) title "Dona 1",1+cos(u)+0.5*cos(u)*cos(v),0.5*sin(v),sin(u)+0.5*sin(u)*cos(v) title "Dona 2"
```

Noten cómo podemos ubicar comentarios con el símbolo $\#\,\rm Comentario$

```gnuplot
reset

set key bmargin right # Abajo del margen - Éste es un comentario

set parametric

set isosamples 20,20

set hidden3d

set title "Donas entrecruzadas"

set xlabel "{/Symbol a}"
set ylabel "{/Symbol b}"
set zlabel "{/Symbol g}"

set urange [-pi:pi]
set vrange [-pi:pi]

#Podemos poner títulos a las leyendas con title "título"

splot cos(u)+0.5*cos(u)*cos(v),sin(u)+0.5*sin(u)*cos(v),0.5*sin(v) title "Dona 1",1+cos(u)+0.5*cos(u)*cos(v),0.5*sin(v),sin(u)+0.5*sin(u)*cos(v) title "Dona 2"
```

```gnuplot
reset

set key bmargin right box 2 # Abajo del margen - Éste es un comentario

set parametric

set isosamples 20,20 # Define iso-superficies

set hidden3d # Oculta líneas no visibles

set title "Donas entrecruzadas"

set xlabel "{/Symbol a}" offset -1,-0.5
set ylabel "{/Symbol b}" offset -2.0,-0.5
set zlabel "{/Symbol g}" offset +1.0,-0.5

set urange [-pi:pi]
set vrange [-pi:pi]

#Podemos poner títulos a las leyendas con title "título"

splot cos(u)+0.5*cos(u)*cos(v),sin(u)+0.5*sin(u)*cos(v),0.5*sin(v) title "Dona 1",1+cos(u)+0.5*cos(u)*cos(v),0.5*sin(v),sin(u)+0.5*sin(u)*cos(v) title "Dona 2"
```

```gnuplot
reset 

set title "Función e^{|x|}"

set xrange[-4:4]
set yrange[0:50]

set key left bottom

set xtics -4,2,4
set ytics 0,10,50

set xlabel "Eje X"
set ylabel "Eje Y" rotate by 0 offset +1,0

plot exp(abs(x)) title " y = e^{|x|}"
```

### TERMINALES Y EXPORTACIÓN DE GRÁFICAS:

Podemos controlar el tipo y tamaño de letra con $\textbf{font ",número"}$  
    
Y los tamaños se especifican de manera usual: 8, 10, 12, 14, 16.

```gnuplot
reset 

set title "Función e^{|x|}" font ",14"

set xrange[-4:4]
set yrange[0:50]

set key left bottom font ",12"

set tics font ",14" #Modifica el tamaño de letra de los ejes. En este caso de los números.
set xtics -4,2,4
set ytics 0,10,50

set xlabel "Eje X" font ",14" offset -6,+1.5
set ylabel "Eje Y" font ",14" rotate by 0 offset +1.5,0

plot exp(abs(x)) title " y = e^{|x|}"
```

Podemos cambiar el tipo de letra definiendo una terminal distinta con $\textbf{set term}$:

```gnuplot
set term
```

```gnuplot
reset 

set term png font "arial,14"

set title "Función e^{|x|}" font ",14"

set xrange[-4:4]
set yrange[0:50]

set key left bottom font ",12"

set tics font ",14" #Modifica el tamaño de letra de los ejes. En este caso de los números.
set xtics -4,2,4
set ytics 0,10,50

set xlabel "Eje X" font ",14" offset -6,+1.5
set ylabel "Eje Y" font ",14" rotate by 0 offset +1.5,0

plot exp(abs(x)) title " y = e^{|x|}"
```
